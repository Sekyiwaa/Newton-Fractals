```python
def newtonfrac2(cmf, cmfprime, z_0, N, tol):
    
    iter_num = 0
    zprev = z_0
    m,n = z_0.shape
    conv_vals = ones((m,n))*-1
    while iter_num <= N:
        z = zprev - cmf(zprev)/cmfprime(zprev)
#        conv_vals[(abs(z - zprev)<tol)] = iter_num
        conv_vals[(abs(z - zprev)<tol) & (conv_vals == -1)] = iter_num
        zprev = z
#        conv_vals[(abs(cmf(z))<tol) & (conv_vals == -1)] = iter_num
        iter_num += 1
#        print(iter_num)
    return conv_vals
    
```

```python
def newtonfrac(cmf, cmfprime, z_0, N, tol):
    #This code finds the root of the function func using
    #Newton's method with initial guess z_0 until the absolute value
    #of func at z_0 is less than tol or the number of iterations exceed N

#    if (cmfprime(z_0) == 0) or (tol < 0):
#        raise ValueError("The function prime must not be zero at z_0 \
#         and tol must be positive")

    i = 0
    zprev = z_0
    z= zprev - cmf(zprev) / cmfprime(zprev)

    while (i <= N):
        i += 1

#        if i > N:
#            print('function exited after %d iterations without converging' % (i))
#            return

        z_0prev = z
        z = z- cmf(z) / cmfprime(z)
#        z_0=x+y*1j

#    print('Function converged to the root %f after %d iterations' % (z, i))
    return z
```

# Color by the Exponential Smoothing -  $ \qquad \sum{ e ^{-\frac{1}{ \left| \; z_{_i} \; - \; z_{_{i-1}} \;\right|}}}$

```python
def newtonfrac2(cmf, cmfprime, z_0, N, tol):
    
    iter_num = 0
    zprev = z_0
    m,n = z_0.shape
    conv_vals = ones((m,n))*-1
    conv_color = zeros((z_0.shape[0],z_0.shape[1]))
    while iter_num <= N:
        z = zprev - cmf(zprev)/cmfprime(zprev)

        conv_vals[(abs(z - zprev)<tol) & (conv_vals == -1)] = conv_color[(abs(z - zprev)<tol) & (conv_vals == -1)]
    

        iter_num += 1
        conv_color += np.exp(-1/abs(zprev - z))
        zprev = z
        

    return (conv_vals)
```
